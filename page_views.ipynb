```python
df100.Pageviews.sum()
```

```python
patterns = ["^/support", "^/faq", "^/systems", "^/get-started", "^/learn", "^/education", "^/services", "^/research", "^/about"]
```

```python
sum_local = 0
sum_global = df100.Pageviews.sum()
for pattern in patterns:
    ct = df100[df100.Pages.str.contains(pattern, regex=True)].Pageviews.sum()
    sum_local += ct
    print(pattern, ct, 100 * ct / df100.Pageviews.sum())
homepage = df100[df100.Pages == "/"].Pageviews.sum()
print ("homepage", homepage, 100 * homepage / sum_global)
other = sum_global - sum_local - homepage
print("other", other, 100 * other / sum_global)
```
